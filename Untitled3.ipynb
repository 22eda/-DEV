```python
import matplotlib.pyplot as plt

yillar = df['yillar']
denizlerde_yetistiricilik = df['denizlerde yetisticilik']
ic_sularda_yetistiricilik = df['ic sularda yetistiricilik']
toplam = df['toplam']

plt.figure(figsize=(10, 6))

plt.plot(yillar, denizlerde_yetistiricilik, label='Denizlerde Yetiştiricilik', marker='*')
plt.plot(yillar, ic_sularda_yetistiricilik, label='İç Sularda Yetiştiricilik', marker='*')
plt.plot(yillar, toplam, label='Toplam', marker='*')


plt.title('Yıllara Göre Yetiştiricilik Miktarları')
plt.xlabel('Yıllar')
plt.ylabel('Yetiştiricilik Miktarı')
plt.legend()

plt.grid(True)
plt.show()
```

```python

data_2022 = df[df['yillar'] == 2022]

labels = ['Denizlerde Yetiştiricilik', 'İç Sularda Yetiştiricilik']
sizes = [data_2022['denizlerde yetisticilik'].values[0], data_2022['ic sularda yetistiricilik'].values[0]]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

plt.title('2022 Yılı Yetiştiricilik Dağılımı')

plt.show()
```

```python

data_matrix = np.array([df['denizlerde yetisticilik'], df['ic sularda yetistiricilik']])

fig, ax = plt.subplots()
im = ax.imshow(data_matrix, cmap='viridis', interpolation='nearest', aspect='auto')

cbar = plt.colorbar(im)
cbar.set_label('Yetiştiricilik Miktarı')

ax.set_xticks(np.arange(len(df['yillar'])))
ax.set_xticklabels(df['yillar'])
ax.set_yticks(np.arange(2))
ax.set_yticklabels(['Denizlerde Yetiştiricilik', 'İç Sularda Yetiştiricilik'])

ax.set_title('Yıllara Göre Yetiştiricilik Miktarları (2D Görselleştirme)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(2, 1, layout='constrained')

dt = 0.01
t = np.arange(0, 30, dt)

# Fixing random state for reproducibility
np.random.seed(19680801)


nse1 = np.random.randn(len(t))                 # white noise 1
nse2 = np.random.randn(len(t))                 # white noise 2
r = np.exp(-t / 0.05)

cnse1 = np.convolve(nse1, r, mode='same') * dt   # colored noise 1
cnse2 = np.convolve(nse2, r, mode='same') * dt   # colored noise 2

# two signals with a coherent part and a random part
s1 = 0.01 * np.sin(2 * np.pi * 10 * t) + cnse1
s2 = 0.01 * np.sin(2 * np.pi * 10 * t) + cnse2

ax1.plot(t, s1, t, s2)
ax1.set_xlim(0, 5)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('s1 and s2')
ax1.grid(True)

cxy, f = ax2.csd(s1, s2, 256, 1. / dt)
ax2.set_ylabel('CSD (dB)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Verileri oluşturalım
data = {
    'Yıl': [2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011,
            2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Balıklandırılan İl Sayısı': [28, 48, 57, 48, 49, 55, 56, 58, 58, 62,
                                  39, 49, 32, 52, 55, 60, 60, 66, 63],
    'Balıklandırılan Kaynak Sayısı': [200, 274, 338, 315, 317, 326, 330, 401,
                                      451, 449, 498, 554, 328, 378, 541, 591,
                                      686, 798, 833],
    'Balık Miktarı (adet)': [5540000, 5920000, 6000000, 6485000, 6550000,
                              4450000, 4730000, 4100000, 5150000, 4550000,
                              3500000, 4250000, 4030000, 3500000, 5000000,
                              6085000, 6000000, 6168000, 6691150]
}

df = pd.DataFrame(data)

# CSD grafiği için s1 ve s2 verilerini oluşturalım
t = df['Yıl']
s1 = df['Balık Miktarı (adet)'].values
s2 = df['Balıklandırılan İl Sayısı'].values

# Zaman adımı
dt = t[1] - t[0]

# CSD hesaplama
cxy, f = plt.csd(s1, s2, NFFT=len(t), Fs=1/dt, scale_by_freq=True)

# Grafik oluşturma
plt.figure(figsize=(10, 6))
plt.semilogy(f, np.abs(cxy), 'r')
plt.title('Cross-Spectral Density (CSD) between Balık Miktarı and Balıklandırılan İl Sayısı')
plt.xlabel('Frequency (Hz)')
plt.ylabel('CSD')
plt.grid(True)
plt.tight_layout()

plt.show()
```
